```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
cars_df=pd.read_csv('/content/Cars93 (8).csv')
print(cars_df)
fig,ax=plt.subplots(2,3)
fig.set_figwidth(10)
fig.set_figheight(7)
fig.suptitle("multiple box plots",fontsize=16)
ax[0][0].boxplot(cars_df["Price"][cars_df["Type"]=="Compact"])
ax[0][0].set_title('Compact')
ax[0][1].boxplot(cars_df["Price"][cars_df["Type"]=="Large"])
ax[0][1].set_title('Large')
ax[0][2].boxplot(cars_df["Price"][cars_df["Type"]=="Midsize"])
ax[0][2].set_title('Midsize')
ax[1][0].boxplot(cars_df["Price"][cars_df["Type"]=="Small"])
ax[1][0].set_title('Small')
ax[1][1].boxplot(cars_df["Price"][cars_df["Type"]=="Sporty"])
ax[1][1].set_title('Sporty')
ax[1][2].boxplot(cars_df["Price"][cars_df["Type"]=="Van"])
ax[1][2].set_title('Van')
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
min_value=np.min(cars_df['Price'])
q1=np.percentile(cars_df['Price'],25)
median=np.median(cars_df['Price'])
q3=np.percentile(cars_df['Price'],75)
max_value=np.max(cars_df['Price'])
print('minimumvalue:',min_value)
print('q1:',q1)
print('median:',median)
print('q3:',q3)
print('maximumvalue:',max_value)
plt.hist(cars_df['MPG.city'],color='black')
plt.xlabel('MPG.city')
plt.title('frequency distribution of mpg.city')
plt.show()
plt.scatter(cars_df["Weight"],cars_df["MPG.city"])
plt.xlabel(cars_df["Weight"])
plt.ylabel("MPG.city")
plt.title("weight vs MPG.city",fontsize=16)
plt.show()
plt.plot(cars_df["EngineSize"],cars_df["Horsepower"])
plt.xlabel("EngineSize")
plt.ylabel("Horsepower")
plt.title('line chart to observe the variation in "EngineSize" against "Horsepower"')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Breast_Cancer.csv')
print(df)
print(df.info())
print(df.isnull().sum())
print(df.describe())
print(df.columns)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Status']=le.fit_transform(df['Status'])
print(df['Status'])
x=df.drop('Status',axis=1)
print(x)
y=df['Status']
print(y)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.23,random_state=0)
print('xtrain shape:',xtrain.shape)
print('xtest:',xtest.shape)
print('ytrain:',ytrain.shape)
print('ytest:',ytest.shape)

# Select categorical columns for one-hot encoding
categorical_cols = x.select_dtypes(include='object').columns

# Apply one-hot encoding to the training and testing dataframes
xtrain_encoded = pd.get_dummies(xtrain, columns=categorical_cols, drop_first=True)
xtest_encoded = pd.get_dummies(xtest, columns=categorical_cols, drop_first=True)

# Align columns between training and testing sets after one-hot encoding
xtrain_encoded, xtest_encoded = xtrain_encoded.align(xtest_encoded, join='left', axis=1, fill_value=0)


from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(random_state=1)
model.fit(xtrain_encoded,ytrain)
train_accurancy=model.score(xtrain_encoded,ytrain)
print(train_accurancy)
test_accruancy=model.score(xtest_encoded,ytest)
print(test_accruancy)
train_predections=model.predict(xtrain_encoded)
print(train_predections)
test_predections=model.predict(xtest_encoded)
print(test_predections)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('/content/Breast_Cancer.csv')
print(df)
print(df.info())
print(df.isnull().sum())
print(df.describe())
print(df.columns)

le = LabelEncoder()
df['Status'] = le.fit_transform(df['Status'])
print(df['Status'])

x = df.drop('Status', axis=1)
y = df['Status']
print(x)
print(y)

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.23, random_state=0)
print('xtrain shape:', xtrain.shape)
print('xtest:', xtest.shape)
print('ytrain:', ytrain.shape)
print('ytest:', ytest.shape)

# Select categorical columns for one-hot encoding
categorical_cols = x.select_dtypes(include='object').columns

# Apply one-hot encoding to the training and testing dataframes
xtrain_encoded = pd.get_dummies(xtrain, columns=categorical_cols, drop_first=True)
xtest_encoded = pd.get_dummies(xtest, columns=categorical_cols, drop_first=True)

# Align columns between training and testing sets after one-hot encoding
xtrain_encoded, xtest_encoded = xtrain_encoded.align(xtest_encoded, join='left', axis=1, fill_value=0)


model = LinearRegression()
model.fit(xtrain_encoded, ytrain)
y_pred = model.predict(xtest_encoded)

print(model.coef_)
print(model.intercept_)

train_score = model.score(xtrain_encoded, ytrain)
print('trainscore(R-sqaured):', train_score)
test_score = model.score(xtest_encoded, ytest)
print('testscore(R-sqaured):', test_score)

mse = mean_squared_error(ytest, y_pred)
mae = mean_absolute_error(ytest, y_pred)
print('mse:', mse)
print('mae:', mae)
```
